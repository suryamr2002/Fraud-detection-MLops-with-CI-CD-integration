```python
id_path = r"C:\Users\SuryaMurugan\Documents\kovai-llm\learning\Data Science Projects\Fraud Detection-mlops\data\ieee-fraud-detection\train_identity.csv"
train_id = pd.read_csv(id_path)
train_id.head()
```

```python
# join the two dataframes on 'TransactionID'
full_df  = pd.merge(train_txn, train_id, on='TransactionID', how='left')
full_df.head()
```

```python
full_df.iloc[0].to_json()
```

```python
full_df["isFraud"] = full_df["isFraud"].astype(float)

# Drop columns with >90% missing values
missing_ratio = full_df.isnull().mean()
cols_to_drop = missing_ratio[missing_ratio > 0.9].index
full_df = full_df.drop(columns=cols_to_drop)

# Fill remaining missing values
full_df = full_df.fillna(0)
```

```python
# take numberi columns
final_df = full_df.select_dtypes(include=["int64", "float64"])
```

```python
final_df.to_parquet("processed_train.parquet", index=False)
print("Feature engineering (pandas) completed.")
```
